```python
train_split = int(len(X_ob) * 0.8)
x_ob_train, x_trade_train, y_train = X_ob[:train_split], X_trade[:train_split], Y_data[:train_split]
x_ob_val, x_trade_val, y_val = X_ob[train_split:], X_trade[train_split:], Y_data[train_split:]
```
